### Nodes

```python
nodes_org = tot[['Organisation', 'org_country', 'org_city_coo']].drop_duplicates()
nodes_org['kind'] = 'Target'
nodes_inv = tot[['Investors', 'inv_country', 'inv_city_coo']].drop_duplicates()
nodes_org = nodes_org.rename(columns={"Organisation": "Node", "org_country": "Country", "org_city_coo": "Coo"})
nodes_inv = nodes_inv.rename(columns={"Investors": "Node","inv_country": "Country", "inv_city_coo": "Coo"})
nodes_inv['kind'] = 'Bidder'
nodes = nodes_org.append(nodes_inv, ignore_index= True)
nodes['id'] = nodes.index
nodes.head()
```

```python
nodes = nodes.assign(coo = nodes.Coo.str.replace('(','').str.replace(')',''))
nodes= nodes.assign(coo = nodes.coo.str.split(', '))

lats = [0]*len(nodes)
longs = [0]*len(nodes)
for i in range(len(nodes)):
    if np.any(pd.isnull(nodes.coo[i])) == True: 
        lats[i] = np.nan
        longs[i] = np.nan
    else :
        lats[i] = nodes.coo[i][0]
        longs[i] = nodes.coo[i][1]

nodes = nodes.assign(lat = lats, long = longs)
nodes.lat = pd.to_numeric(nodes.lat, errors='coerce')
nodes.long = pd.to_numeric(nodes.long, errors = 'coerce')
```

```python
tot3=pd.merge(tot, nodes[nodes['kind']== 'Target'],  how='left', left_on=['Organisation','org_country'], right_on = ['Node','Country'])
tot3.head()
```

```python
tot2 = pd.merge(tot, nodes[nodes['kind']== 'Target'],  how='left', left_on=['Organisation','org_city_coo'], right_on = ['Node','Coo'])
tot2 = pd.merge(tot2, nodes[nodes['kind']== 'Bidder'],  how='left', left_on=['Investors','inv_city_coo'], right_on = ['Node','Coo'])

dic = {}
for i in range(len(nodes)):
               dic[i] = {"Name": nodes.Node[i], 'Country':nodes.Country[i], 'kind': nodes.kind[i]}
```

```python
Pallottola = tot2[tot2.columns[~tot2.columns.str.contains('Unnamed:')]]
```

```python
IDs=Pallottola[['Funding type', 'Node_x']].drop_duplicates()
IDs['idx']=IDs.index
```

```python
Pallottola = pd.merge(Pallottola, IDs,  how='left', left_on=['Funding type','Node_x'], right_on = ['Funding type','Node_x'])
Pallottola.head()
```

```python
Pallottola.id_x
```

```python
bidder_nodes = nodes[nodes['kind'] == 'Bidder'].index
dic_bidder={}
for i in bidder_nodes:
    dic_bidder[i]=dic[i]
```

### Links
*Alert*: the following for loops take quite a while

First way: classical links

```python
links_f = []
for i in Pallottola['idx'].unique():
    for couple in itertools.combinations(np.unique(Pallottola[Pallottola['idx']==i].id_y.values),2):
            links_f.append((couple + ({'m_raised': sum(np.unique(Pallottola[Pallottola['idx']==i]['USDraised_cpi'])),
                                       # it SHOULD be ok as we are not able to check if the same investor, in the same round, 
                                       # in the same organisation but in two different dates has invested the same amount of money
                             'date': min(pd.to_datetime(Pallottola[Pallottola['idx']==i]['Date'].values)),
                             'round': Pallottola[Pallottola['idx']==i]['Funding type'].values[0],
                             'number': 1/(len(Pallottola[Pallottola['idx']==i]['Node_y'].unique())),
                             'type': 'invinv'},)))
```

Second way: including self-loops

```python
links_f_self_loop = []
for i in Pallottola['idx'].unique():
    for couple in itertools.combinations_with_replacement(np.unique(Pallottola[Pallottola['idx']==i].id_y.values),2):
            links_f_self_loop.append((couple + ({'m_raised': sum(np.unique(Pallottola[Pallottola['idx']==i]['USDraised_cpi'])),
                                       # it SHOULD be ok as we are not able to check if the same investor, in the same round, 
                                       # in the same organisation but in two different dates has invested the same amount of money
                             'date': min(pd.to_datetime(Pallottola[Pallottola['idx']==i]['Date'].values)),
                             'round': Pallottola[Pallottola['idx']==i]['Funding type'].values[0],
                             'number': 1/(len(Pallottola[Pallottola['idx']==i]['Node_y'].unique())),
                             'type': 'invinv'},)))
```

Third way: follow your stars

```python
G_bidder_follow=nx.MultiDiGraph()
for org in Pallottola.Organisation.unique():
    ord_list = list(Pallottola[Pallottola.Organisation==org].sort_values(by=['Date']).id_y)
    time_list= list(Pallottola[Pallottola.Organisation==org].sort_values(by=['Date']).Date.dt.year)
    for idx in range(len(ord_list)):
        if ord_list[idx]!= np.nan:
            for seq in range(idx, len(ord_list)):
                if ord_list[seq]!=np.nan:
                    G_bidder_follow.add_edge(ord_list[idx], ord_list[seq], year=time_list[seq])
```
